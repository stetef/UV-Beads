## UV beads have nanoparticles that turn UV (ultraviolet) light into visible light

UV beads contain nanoparticles that *change shape* when exposed to UV light. Their new shape allows them get larger. However, they do not like this larger (excited) state. To go back to their small, normal shape, the nanoparticles will absorb heat and shift back into their preferred smaller state. When they shift back, they will release light in the visible spectrum. This exchange is due to **conservation of energy.**

<br/><br/>
**(Advanced) Details of the chemistry:**

<img src="Figures/uv_bead_dye_structure.svg" width="400">

The nanoparticles in UV Beads consist of two largely carbon systems joined together. Because each system is mostly double-bonded carbons, they form flat ("planar") rings. Organic carbon compounds are often found in this shape. (By convention, people draw carbon structures like the above figure, where the carbons are assumed to be at each corner of the ring.)

The bond that combines these two systems is shown by the orange carbon-to-oxygen bond in the figure above.

*The important fact is that no resonance occurs between the two systems in this state because it is too small to interact with UV light. Thus, it is initially colorless.*

When UV light shines on the beads, the nanoparticle gets "excited" and gains energy, like you would with a sugar rush. The orange bond breaks, allowing the two systems to come apart and the entire molecule to expand in size. The larger the size of the molecule, the longer the wavelength of light that can be emitted by it. By changing the size of the two systems of the molecule, one can achieve different colors.

Heat from the bead in the surrounding area provides enough activation energy for the nanoparticle to return to its original, colorless structure (which is lower in energy). Hence, the larger structure is called its "excited" state, while the original, low energy structure is called its "ground" state.

Each company that makes UV beads can have its own formulation (a trade secret) because multiple structures can behave this way. This variation in structures that have the same property is what allows for multiple colored beads.

**Super advanced:** The planar rings of double-bonded carbon are called *conjugated* because there is an extra electron that gets dispersed around the entire system of double-bonded carbons. This special electronic configuration allows for *resonances* to be formed among all the double-bonded carbons. It is this conjugation that makes the system special and interact with UV light like it does. 

Carbon likes to participate in conjugated systems (alternating single and double-bonded carbons) because it is lower in energy (and thus easier to form)!

## The electromagnetic spectrum

The scientific definition of light is *electromagnetic radiation.*

To explain what this means, ket's first talk about the idea of **particle-wave duality** for light.

This term means that sometimes light can behave like a particle (light particles are called *photons*) and sometimes it behaves like a wave, depending on what experiment you are performing, or how you observe it.

**Advanced:** *Quantum mechanics* states light is both a particle and wave at the same time! In fact, a famous French physicist named **Louis de Broglie** came up with a formula stating that *everything* is both a wave and particle, but the more mass (matter) you have, the more you behave like a particle rather than a wave. Light is special because it doesn't have mass.

<img src="Figures/wave_particle.svg" width="400">

<br/><br/>
**Waves**

The wave part of light is important here. Waves are defined by certain characteristics. Like a wave that crashes onto the beach, the size or strength of the wave is determined by the height of the wave, which we call *amplitude*.

Another important characteristic of waves is that it is periodic, or it repeats. Just like ripples in water, there are alternating high and low points.

The distance between high points, called peaks or crests, in the wave is called its **wavelength**. We often denote wavelength with the Greek letter "lambda": $\lambda$.

For light, this wavelength determines how much **energy** the light contains.

See the figure below for a diagram of characteristic parts of a wave.

# Calculation 1

We can use the speed of a wave to calculate its frequency, which is defined as 
$f = \frac{v}{\lambda}$

Here $v$ is the speed of the wave (in meters over seconds, $\frac{m}{s}$) and $\lambda$ is its wavelength (in meters, m). The units of frequency are inverse seconds ($\frac{1}{s}$), which we give a new name -- *Hertz* -- which we denote as Hz.

<img src="Figures/lightspeed.jpg" width="400">

The speed of light (in the vaccum  of space) is a universal constant (while we dontae as the letter c, which is equal to $3 * 10^8 \frac{m}{s}$). This means that nothing can travel faster than it.

Radio waves have wavelengths between 1 mm and 10,000 km. The different wavlenegths correspond to different frequencies, which is what you tune to when selecting different radio stations in your car.

**4G versus 5G**

So why do we use radio waves for, well, the radio? Why can't we use X-rays to send our favorite music artist across the city? This phenomenon is the same reason why your Bluetooth headphones can't connect when far away from your phone and why Wi-Fi is so local. It has to do with the inverse relationship between the energy of a wave and the distance a wave can travel.

High energy waves (small wavelength and large frequency) like X-rays interact with *things* (or matter) much more, whether that be mountains, buildings, or the bones that you're getting X-rayed at the doctor's office. Because they interact more, they can only travel short distances. This phenomenon is called *attenuation*. However, low energy waves (longer wavelength and small frequency) like radio waves don't interact with many things that often, allowing them to travel farther distances because of their small attenuation.

4G (fourth generation) cellular data is in the range 600 MHz (mega-Hz, which is $10^6$ Hz) to 2.5 GHz (giga-Hz, which is $10^9$ Hz). This range is lower in frequency (and thus lower in energy and larger in wavelength) than 5G (5th generation). 

*What does that mean?* 5G can have faster connection, but cellular towers need to be closer. In this case, 5G towers need to be within line-of-sight of receivers (like your phone). That is why cellular companies built more towers before releasing 5G phones on the market.

You can connect atoms in a line (to make your one-dimensional or 1D nanoparticle) by placing them on *lattice points*. A lattice is a grid of regularly spaced lines where atoms can go. Often, scientists discuss lattices in the context of crystals, or any structure with regularly repeating atoms.

For example, diamonds are crystals because they have repeating units of carbon all bonded together. Can you think of other crystals? See the structure of diamond below.

In this case, let's assume light will interact, or resonate, with all sub-chains within your lattice.

For example, if you have three atoms bonded together, light will interact with both bonds (both pairs on atoms on either side of the central atom) *and* it will interact with the entire length of the atom chain, where you would add both bond lengths together to get the size of the largest interaction.

Likewise, light will interact with all sub-chains. So if a chain is composed of five atoms, it will iteract with the entire chain (all five atoms over all four bonds) and it will interact with all bonds that make up four-atom chains (atoms 1 to 4 and atoms 2 to 5).

Another sub-chain is the all three atom (two bond) sets. For example, atoms 1 to 3, 2 to 5, 3 to 6.

While light can also interact with a single atom, the wavelength is so small for that type of interaction that we will ignore it for now. Therefore, the light in our region of interest will only interact with bonded atoms.

The distance of each chain will make a resonance, just like with the UV beads. Those distances will determine the wavelength of light that can interact that chain.

The nanoparticle's total color will be an addition of all the resonant wavelengths (called Beer's Law). Here, we have modelled light to interact with the distance between atoms such that every 100 nm of light interacts with one angstrom. Thus, for every bond, or chain of bonds, multiply by 100 to get the nm of light it will produce.

An angstrom is 10^-10 meters, while a nanometer is 10^-9 meters. Atom sizes are on the order of one angstrom. 

Thus, a 250 nm wavelength of light would be emitted from a 2.5 angstrom bond. Or a 5 angstrom chain of bonds would produce 500 nm light.

For example, for the two nanoparticles below, the nanoparticle on the left (made out of three Atom 1) will interact with 210 nm light **and** 420 nm light, with twice as much 210 nm light as 420 nm light.

Similarly, the nanoparticle on the right (made out of three Atom 2) will interact with 250 nm light **and** 500 nm light, with twice as much 250 nm light as 500 nm light.

How to convert bond length (or chain of bonds) to light wavelength:

    1 nm bond or chain of bonds --> multiply by 100 = 100 nm of light

    2 nm bond or chain of bonds --> multiply by 100 = 200 nm of light

    3 nm bond or chain of bonds --> multiply by 100 = 300 nm of light

etc.

## Nanostructures impact macro-properties (but sometimes they don't)

Here are some possible **nanostructure properties** that scientists can change:

1. Size
2. Shape
3. Arrangement
4. Composition

Specifically, most scientists are interested in how these properties affect the material's **macro-properties**, such as:

1. Interaction with light
2. Magnetism/ magnetic susceptibility
3. Electrical conductance
4. Electrical resistance/ superconductivity

Sometimes nanostructures impact macroscopic properties, but sometimes they don't. This fact makes it difficult to predict what nanoscale properties will yield the desired macro-effect and hence why researchers are needed to explore this space.
